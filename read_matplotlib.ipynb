# stacked bar graph

```python
x=["usa","germany","australia","india","russia"]
gold=[40,30,45,3,34]
silver=[100,90,129,54,190]
bronze=[90,40,120,99,70]
plt.bar(x,gold,width=0.4,label="gold")
plt.bar(x,silver,width=0.4,bottom=gold,label="silver")
plt.bar(x,bronze,width=0.4,bottom=silver,label="bronze")
plt.legend()
plt.show()
```
